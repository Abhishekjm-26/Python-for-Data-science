```python
def mf1(r1,r2):
    print(r1,r2)
mf1(100,200)
```

```python
def mf5(args): #very important variable
    print(args)
    print(type(args))
mf5(100)
```

```python
def mf5(*args): #very important object that is tuple
    print(args)
    print(type(args))
mf5(100,'Hello', 200)
```

```python
def mf6(r1,r2,*args1):
    print(r1,r2)
    print(args1)
mf6(10,20,30,40,'Hello')
```

```python
def mf7(a,b,kwargs): #regular parameter like args
    print(a,b)
    print(kwargs)
mf7(10,20,'hello')
```

```python
def mf8(a,b,**kwargs): #It's a dictionary
    print(a,b)
    print(kwargs)
    print(type(kwargs))
mf8(1,2,c = 'hello', d = 'hi')
```

```python
def mf9(a,b,**kwargs):
    print(a,b)
    print(kwargs)
    print(type(kwargs))
    if 'c' in kwargs:
        if kwargs['c'] == 40:
            print('hello')
mf9(20,30,c = 20,d = 50)
```

```python
def getsum(*args):
    s = 0
    for n in args:
        s = s+n
    return s
getsum(20,40,60)
```

# Lambda function

```python
def isEven(num):
    return (num%2 ==0)
print(isEven(21))
```

```python
# lambda syntax
isEven = lambda p:(p%2==0)
print(isEven(42))
```

```python
isEvenOdd = lambda p0 : "Even" if (p0%2)==0 else "Odd"
print(isEvenOdd(42))  #  "Even"
print(isEvenOdd(43))  #  "Odd"
```

```python
getMultiples = lambda  m: [m*n   for n in range(1,11) ]
print(getMultiples(3)) # [3,6,9,...30]
print(getMultiples(4))
```

```python
def get_3_multiples(): #just list
    olist = []
    while True:
        num = int(input("Enter the numbers: "))
        olist.append(num)
        if len(olist) == 3:
            break

    for i in olist:
        ilist = []
        for n in range(1, 4):
            ilist.append(i * n)
        print(ilist)

    return olist

get_3_multiples()
```

```python
def get_3_multiples(): #nested
    olist = []
    while len(olist) <3:
        num = int(input("Enter the numbers: "))
        olist.append(num)
    result = []
    for i in olist:
        ilist = []
        for n in range(1,4):
            ilist.append(i*n)
        result.append(ilist)
    return result      
get_3_multiples()
```

```python
def get3Multiples_1(mpc):
    olist = []

    for m in mpc:
        ilist = []
        for n in range(1,4):
            ilist.append(m*n)
        olist.append(ilist)

    return olist

get3Multiples_1([3,4,5])
```

```python
get3multiples = lambda mpc : [[  m*n for n in range(1,4)] for m in mpc ]
print(get3multiples([3,4,5]))   
```

```python
a = [1, 2, 3]
b = [4, 5, 6]
c = [7, 8, 9]

def sum_list(a, b, c):
    z = []
    for i in range(len(a)):
        z.append(a[i] + b[i] + c[i])
    return z

print(sum_list(a, b, c))
```

```python
a = [1, 2, 3]
b = [4, 5, 6]
c = [7, 8, 9]
sum_list = lambda : [ a[i] + b[i] + c[i] for i in range(len(a)) ]
print(sum_list())
```

```python
a = [1, 2, 3]
b = [4, 5, 6]
c = [7, 8, 9]

result = map(lambda x, y, z: x + y + z, a, b, c)

print(list(result))
```

# Filter

```python
ilist = list(range(1,21))
ilist
```

```python
fobj = filter(lambda m: (m%2) ==0, ilist)
print(fobj)
print(type(fobj))
print(list(fobj)) #Can be viewed only once
```

```python
fovj = filter(lambda char : char.lower() in 'aeiou', 'pythondatascience')
print(set(fovj))
```

```python
def filter1(container):
    rlist = []
    for char in container:
        if char.lower() in "aeiou":
            rlist.append(char)

    return rlist
filter1("pythondatascience")
```

## Global Scope


```python
x = 20   # global variable

def test():
    print(x)

test()
print(x)
```

## Modifying Global

```python
x = 20

def test():
    x = x + 1   # Error
    return x
print(test())
```

```python
x = 20

def test():
    global x #define explicitly by mentioning it
    x = x + 1  
    return x
print(test())
```

## Nested functions

```python
# Modifications
def outer():
    x = 10

    def inner():
        nonlocal x
        x = 20

    inner()
    print(x)

outer()
```
