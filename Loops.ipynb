# while

```python
# while loop
n1 = 10
n2 = 40
while n1<n2:
    print(n1,end = ',')
    n1+=2
print("Done")
```

```python
ms1 = "Abhishek" #print elements of string using while
i = 0
while i< len(ms1):
    print(ms1[i], end = ',')
    i+=1
```

```python
ms2 = [1,2,3,4,5] #pribt elements of list using while
i = 0
while i<len(ms2):
    print(i, end = ' ')
    i+=1
```

# Comprehension

```python
slist = []
for num in range(1,11):
    r = num ** 2
    slist.append(r)
print(slist)
```

```python
slist = [ num**2 for num in range(1,11)] #single line
print(slist)
```

```python
vlist = []

for char in "pythondatascience":
    if char in "aeiouAEIOU":
        r = char
        vlist.append(r)
print(vlist)
```

```python
print( [ char for char in "pythondatascience" if char in "aeiouAEIOU" ] )
```

```python
oelist3 = []

for n in range(1,20):
    if n%4 == 0:
        if n%2 == 0:
            r = n**2
        else:
            r = n**3
        oelist3.append(r)
print(oelist3)
```

```python
print([ n**2 if n%2 == 0 else n**3 for n in range(1,20) if n%3 == 0])
```

```python

ulist = ['u1','u2','u3']
plist = ['p1','p2','p3']
uplist1 = []
for u in ulist:
    if u != 'u2':
        for p in plist:
            if p!='p2':
                r = (u,p)
                uplist1.append(r)
print(uplist1)
```

```python
ulist = ['u1','u2','u3'] #cartersian product
plist = ['p1','p2','p3']
uplist = []
for u in ulist:
    if u == ulist[0]:
        for p in plist:
            r = (u,p)
            uplist.append(r)
print(uplist)
```

```python
ulist = ['u1','u2','u3'] 
plist = ['p1','p2','p3']
print([(u,p)   for u in ulist    if u == ulist[0]       for p in plist ])
```

```python
rlist = [n%3  for n in range(1,20) ]
print(rlist)
slist = {n%3  for n in range(1,20) }
print(slist)
```

```python
md1 = {}
for u in range(1,20):
    r = u**2
    md1.update({u:r})
print(md1)
```

```python
#md1= {1:2,2:4,3:6....}
md1 = { n:n**2 for n in range(1,20) }
print(md1)
```

# break and while

```python
lis = []
while True:
    item = input("enter item")
    if item.lower() == "endshopping":
        break #comes out if condition is met and where write condition first and increment next for iteration is required in while
    lis.append(item)
print(lis)
    
```

```python
for ch in "Python":
    if ch == 'o':
        continue #skips when condition met and write increment first if iteration is required in while and condition next
    print(ch, end = '')
```
