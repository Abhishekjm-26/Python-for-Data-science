```python
ms1 = "hello world"
ms1.find('o')
ms1.find('w',0)
ms1.find('w',0,7)
print(ms1.find("hello")) # returns lowest substring
```

```python
ms2 = 'pdshJik hdjd'
ms2.count('S',3) #similiar to find
ms2.lower()
ms2.upper()
ms2.swapcase()
ms2.title() # first letter of each paragraph making all other letters small. 
ms2.capitalize() # Just capitalizes first letter of the string or paragraph making rest small.
```

```python
ms3 = "abhi8"
ms3.isalpha()
ms3.isalnum()
ms3.isupper()
ms3.islower()
```

```python
ms4 = "Abhishek_works_at_Google" #split string to list
slist = ms4.split('_',1)
blist = ms4.split()
print(slist)
print(blist)
print(len(slist))
print(len(blist))
```

# Formatting

```python
help(str.format)
```

```python
ns1 = "Average {} marks of {} students is {}".format('Math', 50,98.9)
ns1
```

```python
ns2 = "Average {:s} marks of {:d} students is {:f}".format('Math', 50,98.9)
ns2
```

```python
#data type constraint
subt = 'math'; marks = 50; percentage = 98.965
ns3 = "Average {:s} marks of {:d} students is {:.3f}".format(subt, marks, percentage) #.2f or.3f is 2 digits after a decimal place
ns3
```

```python
#replacement of index with data type constraint
ns4 = "Average {1:f} marks of {2:f} students is {0:s}".format('Math', 50,98.9) #position start with 0 itself.
ns4
```
