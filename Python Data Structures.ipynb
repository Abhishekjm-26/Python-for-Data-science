# Copy method

```python
m1 = [11,22,33,44,[1,2,3],55]
m2 = m1
m1.append(66)
print(m1)
print(m2)
```

```python
m1 = [1,2,3,4,5,[6,7,8]]
m2 = m1.copy()
m1[5].append(41)
m1.append(44)
print(m1)
print(m2)
```
